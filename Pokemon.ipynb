## How much of each type is there?

```python
counts = []
for type in types:
    n = len(type.index)
    counts.append(n)

#set up a dataframe for types and their population count
type_counts = pd.DataFrame()
type_counts['type'] = type_names
type_counts['count'] = counts

type_counts_by_count = type_counts.sort_values(by='count', ascending=False)

plt.figure(figsize=(15,5))
barlist =plt.bar(type_counts_by_count['type'], type_counts_by_count['count'])
#barlist[0].set_color('r')
#repeat to change colors for all the bars
plt.show()

#print(type_counts_by_count['type'])
print(type_counts_by_count.to_string(index=False))
```

As we can see, there are more much more water pokemon than any other type.  There are nearly 4 times as many water pokemon as there are ice pokemon. This scattered distribution has a strong effect on type balance.

Consider this, pokemon whose type(s) is weak to water based attacks are subtlely weaker, because any one pokemon
launching an attack against them is more likely to be water than any other type and thus, know a water move to use against them. 
Conversely, pokemon who are weak to ice based attacks are subtlely stronger, because any one pokemon launching an attack against them is less likely to be ice. 





## Type Balance Insights


How can we start applying the type distribution of pokemon to balance the types? We can begin to answer that question by looking at how an attack of any type hits the average pokemon.

```python
modifier_mean = []
for type in type_names:
    m_mean = np.mean(pokemon['against_{}'.format(type)])
    modifier_mean.append(m_mean)

type_counts['mean_modifier'] = modifier_mean

type_counts_by_mean = type_counts.sort_values(by='mean_modifier', ascending=False)

plt.figure(figsize=(15,5))
barlist =plt.bar(type_counts_by_mean['type'], type_counts_by_mean['mean_modifier'])
plt.title('Attack Damage Modifier for a Given Type against an Average Pokemon')
plt.xlabel('Attacking Type')
plt.ylabel('Average Damage Modifier')
plt.ylim(.7,1.3)
#barlist[0].set_color('r')
#repeat to change colors for all the bars
plt.show()
```
